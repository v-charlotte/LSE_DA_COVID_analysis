```python
# Create a DataFrame with the text only
tweets_text = tweets[["text"]]
tweets_text = tweets_text.dropna(how='any',axis=0)
```

```python
# Loop through the messages and build a list of values containing the #-symbol
tweets_hashtags = tweets_text[tweets_text["text"].str.contains('#')]
```

```python
# Filter and sort
list_hashtags = []
tweets_hashtags.apply(lambda tweet: list_hashtags.extend([word.strip().strip('.').strip(',').strip('!').strip(':').lower() for word in tweet["text"].split() if word.startswith('#')]), axis=1)

#creating our series in the form '#hashtag' => count
series_keys = list(set(list_hashtags)) #unique values only
series_values = [list_hashtags.count(hashtag) for hashtag in series_keys]
series_hashtags = pd.Series(series_values, series_keys);
print(series_hashtags.sort_values(ascending=False))
```

```python
# Plot
hashtags = pd.DataFrame(series_hashtags)
hashtags = hashtags.rename(columns = {0:'Count'})
hashtags.info()
hashtags.nlargest(n=30, columns=['Count']).plot.barh()

covid_terms = ["covid", "corona", "vaccine", "omicron", "health", "pandemic", "outbreak"]
covid_hashtags = hashtags[[bool([1 for search in covid_terms if search in x]) for x in hashtags.index]]

top_covid_hashtags = covid_hashtags.nlargest(n=30, columns=['Count'])
fig = top_covid_hashtags.plot.barh()

for i, v in enumerate(top_covid_hashtags['Count']):
    fig.text(v, i, str(v), va='center')
plt.savefig("5.png", bbox_inches='tight') # fix layout and save
```

```python
# Define function to plot moving averages
def plot_moving_average(series, window, plot_intervals=False, scale=1.96):
    
    # Create a rolling window to calculate the rolling mean using the series.rolling function
    rolling_mean = series.rolling(window=window).mean()
    
    # Declare the dimensions for the plot, plot name and plot the data consisting of the rolling mean from above 
    plt.figure(figsize=(18,4))
    plt.title('Moving average\n window size = {}'.format(window))
    plt.plot(rolling_mean, 'g', label='Simple moving average trend')

    
    # Plot confidence intervals for smoothed values
    if plot_intervals:
        
        # Calculate the mean absolute square 
        mae = mean_absolute_error(series[window:], rolling_mean[window:])
        
        # Calculate the standard deviation using numpy's std function
        deviation = np.std(series[window:] - rolling_mean[window:])
        
        # Calculate the upper and lower bounds 
        lower_bound = rolling_mean - (mae + scale * deviation)
        upper_bound = rolling_mean + (mae + scale * deviation)
        
        # Name and style upper and lower bound lines and labels 
        plt.plot(upper_bound, 'r--', label='Upper bound / Lower bound')
        plt.plot(lower_bound, 'r--')
    
    # Plot the actual values for the entire timeframe
    plt.plot(series[window:], label='Actual values')
    plt.grid(True)
```

```python
# Define function to calculate the mean absolute error
def mean_absolute_error(a, b): return abs(b - a)
```
